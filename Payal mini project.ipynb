```python
# Feature Engineering and Feature Scaling
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
categorical = list(x_train.select_dtypes('object').columns)
print(f"Categorical columns are: {categorical}")

# Define numerical columns
numerical = list(x_train.select_dtypes('number').columns)
print(f"Numerical columns are: {numerical}")
```

```python
# Define categorical pipeline
cat_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

# Define numerical pipeline
num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler())
])

# Fit column transformer to training data
preprocessor = ColumnTransformer([
    ('cat', cat_pipe, categorical),
    ('num', num_pipe, numerical)
])
preprocessor.fit(x_train)

# Prepare column names
cat_columns = preprocessor.named_transformers_['cat']['encoder'].get_feature_names(categorical)
columns = np.append(cat_columns, numerical)

# Inspect training data before and after
print("******************** Training data ********************")
display(x_train)
x1_train=pd.DataFrame(preprocessor.transform(x_train), columns=columns)

# Inspect test data before and after
print("******************** Test data ********************")
display(x_test)
x1_test=pd.DataFrame(preprocessor.transform(x_test), columns=columns)
```

```python
# Principal Component Analysis
```

```python
from sklearn.decomposition import PCA
 
pca = PCA(n_components = 2)
 
x1_train = pca.fit_transform(x1_train)
x1_test = pca.transform(x1_test)
 
explained_variance = pca.explained_variance_ratio_
```

```python
# Modelling on the dataset
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
classifier=DecisionTreeClassifier()
reg=LogisticRegression()
```

```python
 def algoapplying(object,x,y):
    try:                
        object.fit(x,y)
        pred1=object.predict(x)
        a=accuracy_score(y,pred1)
        
        return a
        
    
    except Exception as e:
        print(e)
        
        
def classificationreport(object,x,y):
    try:
        object.fit(x,y)
        pred1=object.predict(x)
        a=classification_report(y,pred1)
        
        print(a)
        
    
    except Exception as e:
        print(e)

    
```

```python
# Generating accuracy score
```

```python
print(f"the training acuracy of decision tree model is {algoapplying(classifier,x1_train,y_train)}.")
print(f"the training acuracy of logistic  model is {algoapplying(reg,x1_train,y_train)}.")
print(f"the testing acuracy of decision tree model is {algoapplying(classifier,x1_test,y_test)}.")
print(f"the testing acuracy of logistic  model is {algoapplying(reg,x1_test,y_test)}.")
```

```python
# Classification Report
```

```python
print(f"the training acuracy of decision tree model is {classificationreport(classifier,x1_train,y_train)}.")
print(f"the training acuracy of logistic  model is {classificationreport(reg,x1_train,y_train)}.")
print(f"the testing acuracy of decision tree model is {classificationreport(classifier,x1_test,y_test)}.")
print(f"the testing acuracy of logistic  model is {classificationreport(reg,x1_test,y_test)}.")
```
